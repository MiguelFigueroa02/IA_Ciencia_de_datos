```python
url_clientes = "https://gist.githubusercontent.com/ahcamachod/d08898e1cc84c7271856e78d9558ca3d/raw/2f7e1fa53711efe6433a5c5ec79948f600591cb9/clientes_zoop.csv"
url_ventas = "https://gist.githubusercontent.com/ahcamachod/966d7162a3c3a2769b8770b35abf1bcc/raw/d79941b3fa46a450f3611589abd8663399c1277b/ventas_zoop.csv"
url_zoop_pay = "https://gist.githubusercontent.com/ahcamachod/ce728231a836a973c0df94a9f03ed871/raw/8db6da43d7cd94fcab937208067d0fceb521b78c/zoop_pay.csv"
```

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
clientes = pd.read_csv(url_clientes)
```

```python
clientes.info()
```

```python
ventas= pd.read_csv(url_ventas)
```

```python
ventas.info()
```

```python
ventas['fecha'] = pd.to_datetime(ventas['fecha'],format='%Y-%m-%d')
```

```python
ventas.info()
```

```python
zoop_pay = pd.read_csv(url_zoop_pay)
```

```python
zoop_pay.info()
```

```python
# 1. Unir los DataFrames 'ventas' y 'clientes' usando 'ID_compra' como clave
# Usamos inner join para quedarnos solo con las compras que están en ambas bases
df = pd.merge(ventas, clientes, on='ID_compra', how='inner')
```

```python
# 2. Reordenar las columnas en la secuencia solicitada
columnas_ordenadas = [
    'ID_compra',
    'fecha',
    'horario',
    'categoria',
    'precio_unitario',
    'cantidad',
    'envio',
    'metodo_pago',
    'ID_cliente',
    'edad',
    'sexo_biologico',
    'ciudad',
    'estado',
    'region',
    'cashback',
    'nota'
]
df = df[columnas_ordenadas]
```

```python
# Mostrar las primeras filas del DataFrame final
df.head()
```

```python
# Unir df_completo con zoop_pay usando 'ID_cliente' como llave
df_completo = pd.merge(df, zoop_pay, on='ID_cliente', how='left')
```

```python
df_completo.info()
```

```python
# Mostrar las primeras filas para verificar el resultado
df_completo.head()
```

Tenemos un DataFrame llamado "df" y queremos construir una visualización que presente la proporción de los(as) usuarios(as) que realizaron el pre-registro en Zoop Pay, la nueva solución de Zoop.

En "df", tenemos la columna "zoop_pay" que contiene estos datos. Construye un gráfico de dona donde el valor "Sí" reciba el color #2a9d8f y "No" el color #edede9.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.

```python
# 1. Contar valores de la columna 'zoop_pay'
zoop_pay_resumen = df_completo['zoop_pay'].value_counts().reset_index()
zoop_pay_resumen.columns = ['zoop_pay', 'cantidad']
```

```python
zoop_pay_resumen
```

```python
# 2. Definir colores personalizados para "Sí" y "No"
colores_personalizados = {
    'Sí': '#2a9d8f',    # Verde Zoop
    'No': '#edede9'     # Beige claro
}
```

```python
import plotly.express as px
```

```python
# 3. Crear gráfico de dona
fig = px.pie(
    zoop_pay_resumen,
    names='zoop_pay',
    values='cantidad',
    hole=0.4,
    title='Proporción de usuarios pre-registrados en Zoop Pay',
    color='zoop_pay',
    color_discrete_map=colores_personalizados
)

# 4. Mostrar porcentajes y etiquetas
fig.update_traces(textinfo='percent+label')

# 5. Mostrar gráfico
fig.show()
```

Tenemos un DataFrame llamado "df" y queremos construir una visualización que muestre la distribución de las edades de los(as) clientes por región.

En "df", tenemos las siguientes columnas:

region: región de la persona que realizó la compra.

edad: edad de la persona que realizó la compra.

Deseamos crear una visualización de boxplot que presente la distribución de estos datos. Coloca las edades en el eje x y las regiones en el eje y.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


```python
# Crear boxplot con edades en el eje X y regiones en el eje Y
fig = px.box(
    df,
    x='edad',
    y='region',
    color='region',
    title='Distribución de edades por región',
    labels={'edad': 'Edad', 'region': 'Región'},
    points='all'  # Muestra puntos individuales además del boxplot
)

# Mostrar gráfico
fig.show()
```
